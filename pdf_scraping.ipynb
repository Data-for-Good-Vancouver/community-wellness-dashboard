## Basic PDF scraper

Used to scrape a pdf of school capacities.

Author: Vivian Jin

```python
import pdfplumber
import re

def replace_spaces_with_underscores(input_string):
    # Use regular expression to find spaces between alphanumeric characters
    # and replace them with underscores
    result = re.sub(r'([a-zA-Z])\s+([a-zA-Z])', r'\1_\2', input_string)
    return result


with pdfplumber.open('appendix-g-operating-capacity-and-capacuty-utilization-2021-and-2031.630cc126210.pdf') as pdf:
    # Create a list to store the rows of data
    data = []

    for idx, page in enumerate(pdf.pages):
        curr_page = pdf.pages[idx]
        text = curr_page.extract_text()
        print(text)

        # Split the text into individual lines
        lines = text.strip().split('\n')

        # Iterate through each line and split it into values
        for line in lines:
            cleaned_line = replace_spaces_with_underscores(line)
            if 'Capacity' in cleaned_line or 'School_Name' in cleaned_line or 'Kindergarten' in cleaned_line:
                continue
            values = cleaned_line.split()
            data.append(values)

print(data[5:-2])
```

```python
import csv

# Define the header and data for your CSV file
header = ['School Name', 'Kindergarten', 'Elementary', 'Secondary', 'Total', '2021_utilization' , '2031_utilization']

# Define the file name
file_name = 'school_capacity_parsed_from_pdf.csv'

# Open the CSV file for writing
with open(file_name, mode='w', newline='') as file:
    writer = csv.writer(file)

    # Write the header
    writer.writerow(header)

    # Write the data row by row
    for row in data[5:-2]:
        writer.writerow(row)

    writer.writerow(["Operating capacities are based on VSB average class sizes. Capacity utilizations are based on total enrolment for all programs."])
    writer.writerow(["Original pdf dated Dec. 30, 2021"])
```
